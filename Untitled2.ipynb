```python
def top_10(self,col1,col2,col3):
    t10=self.nlargest(10, col2)[[col1,col2,col3]].reset_index(drop=True)
    return t10
```

```python
top_10_lit_cities=top_10(indian_cities,"name_of_city","effective_literacy_rate_total","state_name")
```
